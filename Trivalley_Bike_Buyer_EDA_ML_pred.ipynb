```python
# ML Code snippet 1
# Setting the noise/redundant/missing data variables
ignored_cols  = ['CustomerID', 'FirstName', 'LastName',  'Phoneumber',
                 'AddressLine1','Title', 'MiddleName', 'Suffix','AddressLine2','PhoneNumber', 'AveMonthSpend','Age']
```

```python
# ML Code snippet 3
# Setting the numerical cols
num_cols = ['YearlyIncome']
```

```python
# ML Code snippet 5
# Installing Pycaret for Automated ML
!pip install -U --pre pycaret
```

```python
# ML Code snippet 6
# Importing Pycaret and setting/configuring the ML experiment
from pycaret.classification import *

regression_model_setup = setup(data,
      target = y,
      categorical_features = cat_cols,
      numeric_features = num_cols,
      ignore_features = ignored_cols)
```

```python
# ML Code snippet 7
# Invoking ML algorithms
compare_models()
```

```python
# ML Code snippet 8
# Creating the model
#If the values of the mean of the results of this instruction match with the results from pycaret, then the model is a good fit
model = create_model('gbc')
```

```python
# ML Code snippet 9
# Importing the client dataset for the prediction
client_data = pandas.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Trivalley%20Client%20data%20file.csv')
client_data.shape
```

```python
#ML Code snippet 9.1
#COnverting the BirthDate to datetime
client_data['BirthDate'] = pandas.to_datetime(client_data['BirthDate'])

```

```python
# ML Code snippet 10
# Predicting 'AveMonthSpend' using the best model
client_data_predictions = predict_model(model, client_data)
```
